Exploratory Data Analysis(EDA)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the visual settings for seaborn
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
# Load the dataset
music_data = pd.read_csv('../data/data_moods.csv')

# Display the first few rows of the dataset
music_data.head()
```

```python
# Basic information about the dataset
music_data.info()

# Check for missing values
music_data.isnull().sum()
```

```python
# Distribution of moods
plt.figure(figsize=(10, 6))
sns.countplot(x='mood', data=music_data, palette='viridis')
plt.title('Distribution of Moods')
plt.xlabel('Mood')
plt.ylabel('Count')
plt.show()
```

```python
# Correlation matrix
correlation_matrix = music_data.corr()

# Plotting the heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Features')
plt.show()
```

```python
# Pair plot of selected features
sns.pairplot(music_data, hue='mood', vars=['popularity', 'energy', 'valence', 'danceability'], palette='viridis')
plt.suptitle('Pair Plot of Selected Features by Mood', y=1.02)
plt.show()
```

```python
# Distribution of energy by mood
plt.figure(figsize=(12, 6))
sns.boxplot(x='mood', y='energy', data=music_data, palette='viridis')
plt.title('Distribution of Energy by Mood')
plt.xlabel('Mood')
plt.ylabel('Energy')
plt.show()
```

```python
# Conclusion
print("Insights:")
print("1. The mood distribution shows that some moods are more prevalent than others, indicating a potential class imbalance.")
print("2. The correlation heatmap reveals that certain features such as energy and loudness are strongly correlated.")
print("3. Pair plots show clear separations for some moods, which could be leveraged for classification.")
```
